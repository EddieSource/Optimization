```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# definition of function f
def f(x, y):
    return (1 - np.cos(np.pi * x / 2)) * (y**2)

# generate 100 x's and 100 y's, zs are their corresponding f values 
xs = np.linspace(-1, 1, 100)
ys = np.linspace(-1, 1, 100)
xys = np.meshgrid(xs, ys)
zs = f(xys[0], xys[1])
```

```python
# set figsize be 7 * 7, plot the contour of f over B
plt.figure(figsize = (7,7))
# set number of levels be 30 so the plot looks smoother
h = plt.contourf(xs, ys, zs, levels = 30)
plt.title('contourf plot of f over B')
plt.axis('scaled')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# compute partial derivative of f over x 
def partialX(x, y):
    return (np.pi / 2) * y**2 * np.sin(np.pi * x / 2)

# compute partial derivative of f over y
def partialY(x, y):
    return 2 * y * (1 - np.cos(np.pi * x / 2))

# make one gradient descent step and return x, y values after this step
def descent(x, y, size):
    x -= size * partialX(x, y)
    y -= size * partialY(x, y)
    return x, y
```

```python
# generate one initial iterates
x_0 = np.random.uniform(-1, 1)
y_0 = np.random.uniform(-1, 1)
```

```python
# make 10 gradient steps
Xs = [x_0]
Ys = [y_0]

for _ in range(10):
    x_0, y_0 = descent(x_0, y_0, 1)
    Xs.append(x_0)
    Ys.append(y_0)
```

```python
# plot the trajectory
plt.figure(figsize = (7,7))
h = plt.contourf(xs, ys, zs, levels = 30)
plt.plot(Xs, Ys, marker='.', linewidth=0.5)
plt.title('a blue trajectory on contourf plot')
plt.axis('scaled')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# use this to get a set of iterates of one trajectory
def get_trajectory():
    x_0 = np.random.uniform(-1, 1)
    y_0 = np.random.uniform(-1, 1)
    Xs = [x_0]
    Ys = [y_0]

    for _ in range(10):
        x_0, y_0 = descent(x_0, y_0, 1)
        Xs.append(x_0)
        Ys.append(y_0)
    return Xs, Ys

# plot 15 trajectories
plt.figure(figsize = (7,7))
h = plt.contourf(xs, ys, zs, levels = 30)
for _ in range(15):
    Xs, Ys = get_trajectory()
    plt.plot(Xs, Ys, marker='.', markersize = 1, linewidth=1)
plt.title('15 trajectories on contourf plot')
plt.axis('scaled')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.figure(figsize = (7,7))
h = plt.contourf(xs, ys, zs, levels = 30)
# plot 1000 trajectories
for _ in range(1000):
    Xs, Ys = get_trajectory()
    plt.plot(Xs, Ys, marker='.', markersize = 0.5, linewidth=0.4)
plt.title('1000 trajectories on contourf plot')
plt.axis('scaled')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
